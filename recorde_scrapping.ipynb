```python
#initiating the driver and linking the given web site website with it
driver = webdriver.Chrome()
driver.get('https://www.giiresearch.com/material_report.shtml')
```

```python
#taking the dates between which the records are required by the user
import datetime
date_entry = input('Enter a initial date in YYYY-MM-DD format')
year, month, day = map(int, date_entry.split('-'))
date1 = datetime.datetime(year, month, day)

date_entry = input('Enter a final date in YYYY-MM-DD format')
year, month, day = map(int, date_entry.split('-'))
date2 = datetime.datetime(year, month, day)
type(date1)
#2022-04-13  2022-04-11
```

```python
#function two calculate the month in number from string
def month_cal(month):
    month=month.lower()
    if month=='january':
        return int(1)
    if month=='febuary':
        return int(2)
    if month=='march':
        return int(3)
    if month=='april':
        return int(4)
    if month=='may':
        return int(5)
    if month=='june':
        return int(6)
    if month=='july':
        return int(7)
    if month=='august':
        return int(8)
    if month=='september':
        return int(9)    
    if month=='october':
        return int(10)
    if month=='november':
        return int(11)
    if month=='december':
        return int(12)
```

```python
#lists where data from records will be storesd
published_by=[]
published=[]
product_code=[]
price=[]
content_info=[]
name=[]
flag=0
```

```python
#here we extract the records from the site and store inn the list above
#here we need to check the conditions for dates and store data according to that
from itertools import count
for j in count(0):
  
  for i in range(len(driver.find_elements_by_class_name('plist_pubinfo'))):
    month=((driver.find_elements_by_class_name('plist_dateinfo')[i].text).split(' '))[0]
    month=(month_cal(month[10:]))
    #print(month)
    if month==None:
        published.append('Pre-Order')
        pb=(driver.find_elements_by_class_name('plist_pubinfo')[i].text)
        published_by.append(pb[13:])
        pd=(driver.find_elements_by_class_name('plist_codeinfo')[i].text)
        product_code.append(pd[13:])
        ci=(driver.find_elements_by_class_name('plist_pageinfo')[i].text)
        content_info.append(ci[12:])
        pr=(driver.find_elements_by_class_name('price_usd')[i].text)
        price.append(price)
        na=(driver.find_elements_by_class_name('plist_t_box')[i].text)
        name.append(na)
    else:
        month=int(month)
        day=((driver.find_elements_by_class_name('plist_dateinfo')[i].text).split(' '))[1]
        day=int(day.split(',')[0])
        year=((driver.find_elements_by_class_name('plist_dateinfo')[i].text).split(' '))[2]
        year=int(year)
        date=(datetime.datetime(year,month,day))
        if date>date2:   #condition 1
          continue
        elif date<date1:  #condition 2
          flag=1
          break
        elif (date>=date1) and (date<=date2): #condition 3
          month=int(month)
          day=((driver.find_elements_by_class_name('plist_dateinfo')[i].text).split(' '))[1]
          day=int(day.split(',')[0])
          year=((driver.find_elements_by_class_name('plist_dateinfo')[i].text).split(' '))[2]
          year=int(year)
          date=(datetime.datetime(year,month,day))
          #print((date))
          published.append(date)
          pb=(driver.find_elements_by_class_name('plist_pubinfo')[i].text)
          published_by.append(pb[13:])
          pd=(driver.find_elements_by_class_name('plist_codeinfo')[i].text)
          product_code.append(pd[13:])
          ci=(driver.find_elements_by_class_name('plist_pageinfo')[i].text)
          content_info.append(ci[12:])
          pr=(driver.find_elements_by_class_name('price_usd')[i].text)
          price.append(pr)
          na=(driver.find_elements_by_class_name('plist_t_box')[i].text)
        name.append(na)
        
  if flag==1:  #condition 4
     break
  if (driver.find_elements_by_class_name('plist_pageinfo')):   #condition 5
         (driver.find_elements_by_class_name('btn_next')[0]).click()
  else:
     break
```

```python
#creating a dataframe for the data
#here I have created a small dataset to demonstrate that the code works perfectly and can scrape large data if required
import pandas as pd
d = {'published_by': published_by, 'publish': published, 'product_code': product_code, 'price': price, 'content_info': content_info, 'name': name}

df = pd.concat([pd.Series(v, name=k) for k, v in d.items()], axis=1)
df
```
